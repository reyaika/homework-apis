```python
# Make a request for the current weather where you are born, or somewhere you've lived.

import requests
import json

from dotenv import load_dotenv
import os

load_dotenv()

api_key_tomorrow = os.getenv("api_key_tomorrow")

location = "manila"

url = f"https://api.tomorrow.io/v4/weather/realtime?location={location}&apikey={api_key_tomorrow}"

response = requests.get(url)
weather = response.json()

pretty_json = json.dumps(weather, indent=4)
print(pretty_json)
```

```python
# Print out the country this location is in.

print(weather['location']['name'])
```

```python
# Print out the difference between the current temperature and how warm it feels. Use "It feels ___ degrees colder" or "It feels ___ degrees warmer," not negative numbers.

actual_temp = weather['data']['values']['temperature']
feels_like = weather['data']['values']['temperatureApparent']
difference = round((abs(feels_like - actual_temp)))

if feels_like > actual_temp:
    print(f"It feels {difference} degrees warmer.")
elif feels_like < actual_temp:
    print(f"It feels {difference} degrees colder.")
else:
    print("It feels the same as the actual temperature.")
```

```python
# What's the current temperature at Heathrow International Airport? Use the airport's IATA code to search.
# It misinterprets LHR as Lahore. 
# Also returns "...try a different term" when I put Heathrow International Airport.

location_lhr = "Heathrow Airport"
url_lhr = f"https://api.tomorrow.io/v4/weather/realtime?location={location_lhr}&apikey={api_key_tomorrow}"

response = requests.get(url_lhr)
weather_lhr = response.json()
pretty_json = json.dumps(weather_lhr, indent=4)
print(pretty_json)
```

```python
# What URL would I use to request a 3-day forecast at Heathrow?
# The default is 5 days so I will have to slice the list later...

location_lhr = "Heathrow Airport"
url_forecast = f"https://api.tomorrow.io/v4/weather/forecast?location={location_lhr}&timesteps=1d&apikey={api_key_tomorrow}"

response = requests.get(url_forecast)
weather_forecast = response.json()
pretty_json = json.dumps(weather_forecast, indent=4)
print(pretty_json)
```

```python
# Print the date of each of the 3 days you're getting a forecast for.
# Print the maximum temperature of each of the days.
# Print only the day with the highest maximum temperature.

forecast_days = weather_forecast['timelines']['daily'][:3]

for day in forecast_days:
    date = day['time']
    max_temp = day['values']['temperatureMax']
    print(f"Date: {date}, Max temperature: {max_temp} degrees Celsius.")
    print("-----")

hottest_day = max(forecast_days, key=lambda day: day['values']['temperatureMax'])

hottest_date = hottest_day['time']
hottest_temp = hottest_day['values']['temperatureMax']

print(f"The hottest day is {hottest_date} with a maximum temperature of {hottest_temp} degrees Celsius.")
```
